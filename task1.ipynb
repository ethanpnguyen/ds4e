## Predicting Tips

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import warnings
warnings.filterwarnings('ignore')
```

```python
tips = pd.read_csv('/Users/ethannguyen/Library/CloudStorage/OneDrive-TheUniversityofTexasatAustin/2022-2023/23 Spring/Data Science/Task 1/tips.csv')
```

```python
tips.describe(include = 'all')
```

```python
x = tips.loc[:, ['total_bill', 'size']]
y = tips.loc[:, 'tip']

reg = LinearRegression()
model = reg.fit(x, y)
```

```python
print(model.intercept_, model.coef_, model.score(x, y))
```

The regression function is:
$$
y = 0.669 + 0.093*total\_bill + 0.193*size
$$

```python
new_total_bill = 45
new_size = 4

print('Predicted tip:', model.predict(np.array([[new_total_bill, new_size]])))
```
